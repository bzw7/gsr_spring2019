## Function for creating custom Databases

```python
head=['id','time_1','lat_1','long_1','heading','speed','consumer_id1','system_time1','x','y','local_time_1',
      'id2','time_2','lat_2','long_2','heading_2','speed_2','consumer_id2','system_time2','x2','y2','local_time_2',
      'speed_m','acceleration', 'trajectory_name']


def db_create(list_columns):
    c=0
    head=['id','time_1','lat_1','long_1','heading','speed','consumer_id1','system_time1','x','y','local_time_1',
      'id2','time_2','lat_2','long_2','heading_2','speed_2','consumer_id2','system_time2','x2','y2','local_time_2',
      'speed_m','acceleration', 'trajectory_name']

    data_dir = 'step_014_probe_add_segment_speed_and_acceleration'
    return_db = pd.DataFrame(columns = head)
    for folder in os.listdir(data_dir):
        file_dir = os.path.join(data_dir, folder)
        file_name = list(filter(lambda x: 'csv' in x, os.listdir(file_dir)))[0]
        file_loc = os.path.join(file_dir, file_name)
        sub_db = pd.read_csv(file_loc, names = head)
        sub_db['trajectory_name'] = file_name
        return_db = pd.concat([return_db, sub_db])
        c+=1
        if c%100==0:
            print (c)
        
    if list_columns == 'all':
        return return_db
    else:
        return return_db[list_columns]
       
```

## Haversine Distance Function

```python
def haversine_dist(lat1, lon1, lat2, lon2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)

    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    meter_dist = 6367000 * c
    return meter_dist
```

```python
col_list = ['id', 'consumer_id1', 'trajectory_name',
            'local_time_1', 'lat_1', 'long_1','x', 'y', 'speed',
            'local_time_2', 'lat_2', 'long_2', 'x2', 'y2', 'speed_2', 
            'speed_m','acceleration']
seg_df = db_create(col_list)
```

```python
rand_traj = np.random.choice(seg_df['trajectory_name'].unique(), 100)
rand_traj
```

```python
rand_traj_df = seg_df[seg_df['trajectory_name'].isin(rand_traj)]
rand_traj_df['local_time_1'] = pd.to_datetime(rand_traj_df['local_time_1'])
rand_traj_df['local_time_2'] = pd.to_datetime(rand_traj_df['local_time_2'])
rand_traj_df.head()
```

```python
rand_traj_df['time_diff'] = (rand_traj_df['local_time_2'] - rand_traj_df['local_time_1']).dt.seconds
rand_traj_df['dist'] = haversine_dist(rand_traj_df['lat_1'], rand_traj_df['long_1'], rand_traj_df['lat_2'], rand_traj_df['long_2'])
rand_traj_df['calc_speed'] = (rand_traj_df['dist'] / rand_traj_df['time_diff']) * 3.6
rand_traj_df['speed_error'] = ((rand_traj_df['speed'] + rand_traj_df['speed_2']) / 2.0) - rand_traj_df['calc_speed']
rand_traj_df.head()
```
